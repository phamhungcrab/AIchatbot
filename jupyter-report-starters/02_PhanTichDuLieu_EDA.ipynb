# Báo cáo EDA

```python

import numpy as np, pandas as pd, matplotlib.pyplot as plt
np.random.seed(42)
data = pd.DataFrame({
    "age": np.random.randint(18, 60, 100),
    "score": np.random.randn(100)*10 + 70
})
data.describe()
```

```python

plt.hist(data["score"], bins=10)
plt.title("Phân phối điểm số")
plt.show()
```
